## 制作一个简单的地图

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

import os
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_map = Basemap(resolution='l', projection='merc', llcrnrlat=10.0, urcrnrlat=55.0, llcrnrlon=60.0, urcrnrlon=140.0)

my_map.drawcoastlines()
plt.show()
```

```python
my_map = Basemap(resolution='l', projection='merc', llcrnrlat=10.0, urcrnrlat=55.0, llcrnrlon=60.0, urcrnrlon=140.0)

my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral', lake_color='aqua')

# 兰色/道奇蓝：DodgerBlue = #1E90FF
#my_map.drawmapboundary(color='k', linewidth=1.0, fill_color='c', zorder=None, ax=None)
my_map.drawmapboundary(color='k', linewidth=1.0, fill_color='#1E90FF', zorder=None, ax=None)

plt.show()
```

```python
my_map = Basemap(resolution='l', projection='merc', llcrnrlat=10.0, urcrnrlat=55.0, llcrnrlon=60.0, urcrnrlon=140.0)

my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral', lake_color='aqua')
my_map.drawmapboundary(color='k', linewidth=1.0, fill_color='#1E90FF', zorder=None, ax=None)

my_map.drawmeridians(np.arange(0, 360, 8))
my_map.drawparallels(np.arange(-90, 90, 6))

plt.show()
```

```python
my_map = Basemap(projection='ortho', lat_0=0, lon_0=-100, resolution='l', area_thresh=1000.0)

my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral', lake_color='aqua')
my_map.drawmapboundary(color='k', linewidth=1.0, fill_color='DodgerBlue', zorder=None, ax=None)

my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))

plt.show()
```

### 设置地图投影

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

import mpl_toolkits.basemap

print(mpl_toolkits.basemap.supported_projections)
```

```python
from mpl_toolkits.basemap import Basemap

import matplotlib.pyplot as plt
import numpy as np

map = Basemap(projection='cyl')
map.drawcoastlines()
map.drawmeridians(np.arange(0, 360, 30))   # 经线
map.drawparallels(np.arange(-90, 90, 30))  # 纬线

plt.show()
```

```python
map = Basemap(projection='aeqd', lon_0=180, lat_0=50)
map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral', lake_color='aqua')
map.drawcoastlines()

plt.show()
```

## 绘制地图背景

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

map = Basemap()
#map = Basemap(llcrnrlon=3.75, llcrnrlat=39.75, urcrnrlon=4.35, urcrnrlat=40.15, epsg=5520)
#map = Basemap(llcrnrlon=72, llcrnrlat=3, urcrnrlon=136, urcrnrlat=54)

map.arcgisimage(service='ESRI_Imagery_World_2D', xpixels=1500, verbose=True)
plt.show()
```

```python
map.arcgisimage(service='World_Shaded_Relief', xpixels=1500, verbose=True)
plt.show()
```
