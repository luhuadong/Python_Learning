## Basemap 基本用法

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
map = Basemap()
```

```python
map.drawcoastlines()
plt.show()
```

```python
map = Basemap(projection='ortho', lat_0=0, lon_0=105)
map.drawcoastlines()
plt.show()
```

```python
map = Basemap(projection='ortho', lat_0=0, lon_0=105)
map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral', lake_color='aqua')
map.drawcoastlines()
map.drawcountries()
plt.show()
```

```python
map = Basemap(projection='ortho', lat_0=0, lon_0=-15)
map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral', lake_color='aqua')
map.drawcoastlines()
map.drawcountries()
plt.show()
```

## 制作一个简单的地图

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

import os
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_map = Basemap(resolution='l', projection='merc', llcrnrlat=10.0, urcrnrlat=55.0, llcrnrlon=60.0, urcrnrlon=140.0)

my_map.drawcoastlines()
plt.show()
```

```python
my_map = Basemap(resolution='l', projection='merc', llcrnrlat=10.0, urcrnrlat=55.0, llcrnrlon=60.0, urcrnrlon=140.0)

my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral', lake_color='aqua')

# 兰色/道奇蓝：DodgerBlue = #1E90FF
#my_map.drawmapboundary(color='k', linewidth=1.0, fill_color='c', zorder=None, ax=None)
my_map.drawmapboundary(color='k', linewidth=1.0, fill_color='#1E90FF', zorder=None, ax=None)

plt.show()
```

```python
my_map = Basemap(resolution='l', projection='merc', llcrnrlat=10.0, urcrnrlat=55.0, llcrnrlon=60.0, urcrnrlon=140.0)

my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral', lake_color='aqua')
my_map.drawmapboundary(color='k', linewidth=1.0, fill_color='#1E90FF', zorder=None, ax=None)

my_map.drawmeridians(np.arange(0, 360, 8))
my_map.drawparallels(np.arange(-90, 90, 6))

plt.show()
```

```python
my_map = Basemap(projection='ortho', lat_0=0, lon_0=-100, resolution='l', area_thresh=1000.0)

my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral', lake_color='aqua')
my_map.drawmapboundary(color='k', linewidth=1.0, fill_color='DodgerBlue', zorder=None, ax=None)

my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))

plt.show()
```
